## Question: Make new column "Total Income = Income.1 + Income.2"

### Solution Date: 14/9/2023

```python
df
```

```python
# Let's try adding both income columns: #WILL FETCH ERROR
df['Total Income'] = df['Income.1'] + df['Income.2']
```

- As the Income.1 contains special char{dollar} so it is treated as object data and we cannot add it to int(Income.2).
- Fix {dollar} sign (remove it).
- Convert Income.1 datatype to float datatype.

```python
# Replace $ with Empty
df['Income.1'] = df['Income.1'].str.replace('$', '')
```

```python
# Replace Empty with np.nan
df['Income.1']= df['Income.1'].replace('',np.nan)
```

```python
print(df.dtypes)
```

```python
# Convert Income.1 to float:

#df['Income.1']= df['Income.1'].astype(float)

df['Income.1'] = pd.to_numeric(df['Income.1'])
```

```python
df.info()
```

```python
# Filling nan's (Imputer Tech can also be used):
df['Income.1'].fillna(df['Income.1'].mean(), inplace=True)
df['Income.2'].fillna(df['Income.2'].mean(), inplace=True)
```

```python
# create a new column "Total Income"(= Income.1+Income.2):
df['Total Income'] = df['Income.1'] + df['Income.2']
```

```python
df['Total Income'] = np.ceil(df['Total Income'])
df
```
